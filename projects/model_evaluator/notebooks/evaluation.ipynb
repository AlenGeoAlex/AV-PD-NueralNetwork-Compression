```python
from datetime import datetime
import os

PATH_TO_MODEL_TO_EVAL = "runs/detect/results/finetune_baseline-26n/yolo26n-ft/weights/best.pt"
MODEL_NAME = "YOLO26n-FT"

TEST_SET_DIRECTORY = "Dataset/merged_dataset/"
```

```python
from datetime import datetime
import os

if not PATH_TO_MODEL_TO_EVAL or not os.path.exists(PATH_TO_MODEL_TO_EVAL):
    raise FileNotFoundError(f"Model path does not exist: {PATH_TO_MODEL_TO_EVAL}")

if not MODEL_NAME or MODEL_NAME.strip() == "":
    raise ValueError("MODEL_NAME cannot be empty")

date = datetime.now().strftime("%Y-%m-%d")
base_output_dir = f"EvaluationResult/{date}/{MODEL_NAME}"

if os.path.exists(base_output_dir):
    time = datetime.now().strftime("%H-%M")
    MODEL_NAME = f"{MODEL_NAME}_{time}"

OUTPUT_DIR = f"EvaluationResult/{date}/{MODEL_NAME}"
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
from eval import run_evaluation

results = run_evaluation(
    model_path    = PATH_TO_MODEL_TO_EVAL,
    test_set_path = TEST_SET_DIRECTORY,
    output_dir    = OUTPUT_DIR,
    model_name    = MODEL_NAME,
)
```
