```python
from dscleaner import DatasetOrchestrator, KITTIReader, EuroCityReader
import logging
logging.basicConfig(level=logging.INFO, format="%(asctime)s  %(levelname)-8s  %(message)s")
```

```python
readers = [
    KITTIReader(
        kitti_root  = "Dataset/KITTI",
        split       = "training",
    ),
    EuroCityReader(
        eurocity_root = "Dataset/EuroCity/Extracted",
        eurocity_label_root = "Dataset/EuroCity/ExtractedVal",
        time_of_day   = "day",
        split         = "val",
    ),
]

orchestrator = DatasetOrchestrator(
    readers     = readers,
    output_dir  = "Dataset/merged_dataset",
    meta_dir    = "Dataset/merged_dataset",
    train_ratio = 0.70,
    val_ratio   = 0.15,
    seed        = 42,
    copy_images = True,
)

orchestrator.run()
```
