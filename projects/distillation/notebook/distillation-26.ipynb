```python
from ft import finetune, FinetuneConfig
from kd.train_kd import main
```

```python
baseline_cfg = FinetuneConfig(
    weights = "yolo26s.pt",
    name    = "yolo26s-ft",
    project = "results/finetune_baseline",
    data    = "Dataset/merged_dataset/dataset.yaml",
    epochs  = 50,
    batch   = 32,
)

student_weights = finetune(baseline_cfg)
print(f"Student weights: {student_weights}")
```

```python
from ultralytics import YOLO

YOLO('yolo26l')

baseline_cfg_teacher = FinetuneConfig(
    weights = "yolo26l.pt",
    name    = "yolo26l-ft",
    project = "results/finetune_baseline-teacher-26",
    data    = "Dataset/merged_dataset/dataset.yaml",
    epochs  = 50,
    batch   = 32,
)

teacher_weights = finetune(baseline_cfg_teacher)
print(f"Teacher weights: {teacher_weights}")
```

```python
from pathlib import Path

student_weights = "runs/detect/results/finetune_baseline/yolo26s-ft/weights/best.pt"
assert Path(student_weights).exists(), f"Weights not found: {student_weights}"
print("Student weights verified")

teacher_weights = "runs/detect/results/finetune_baseline-teacher-26/yolo26l-ft/weights/best.pt"
assert Path(teacher_weights).exists(), f"Weights not found: {student_weights}"
print("Teacher weights verified")
```

```python
results = []

for alpha in [0.3, 0.5, 0.7, 0.9]:
    for temp in [2.0, 3.0, 5.0]:
        print(f"\n{'='*50}")
        print(f"Running KD — alpha={alpha}, temp={temp}")
        print(f"{'='*50}\n")

        main(
            teacher_weights = "yolo26x.pt",
            student_weights = student_weights,
            data_config     = "pedestrian.yaml",
            alpha           = alpha,
            temperature     = temp,
            batch_size      = 32,
            epochs          = 15,
            project_dir     = "runs/kd/grid-search-v26/",
            workers         = 1
        )

        results.append({"alpha": alpha, "temp": temp})

print("\n✅ Grid search complete")
print(f"   Ran {len(results)} combinations")
```
