```python
import zipfile, os, torch
from ultralytics import YOLO

# Unzip YOLO11 40%
with zipfile.ZipFile("/home/jj7/Scalable/yolo11s_pruned_40.pt.zip", 'r') as z:
    z.extractall("/home/jj7/Scalable/yolo11s_40_extracted")

ckpt = torch.load("/home/jj7/Scalable/yolo11s_pruned_40.pt.zip",
                  weights_only=False, map_location="cpu")
torch.save(ckpt, "/home/jj7/Scalable/yolo11s_pruned_40.pt")

model11_40 = YOLO("/home/jj7/Scalable/yolo11s_pruned_40.pt")
model26_40 = YOLO("/home/jj7/Scalable/YOLO26_pruned_40_fineTuned.pt")

print("YOLO11 40% loaded ✓")
print("YOLO26 40% fine tuned loaded ✓")
```

```python
# FP16 engines
model11_40.export(format="engine", imgsz=640, half=True, device=0, batch=1)
model26_40.export(format="engine", imgsz=640, half=True, device=0, batch=1)

# INT8 engines
model11_40.export(format="engine", imgsz=640, int8=True,
                  data="/tmp/calib_data.yaml", device=0, batch=1)
model26_40.export(format="engine", imgsz=640, int8=True,
                  data="/tmp/calib_data.yaml", device=0, batch=1)

print("✅ All 40% engines exported!")
```

```python
import time, glob
from ultralytics import YOLO

def benchmark(path, runs=200):
    m = YOLO(path, task="detect")
    img = glob.glob("/tmp/coco128/images/train2017/*.jpg")[0]
    for _ in range(20): m(img, verbose=False)
    t = time.perf_counter()
    for _ in range(runs): m(img, verbose=False)
    return (time.perf_counter() - t) / runs * 1000

print("Benchmarking 40% models...")
r = {}
r["YOLO11 40% PyTorch"] = benchmark("/home/jj7/Scalable/yolo11s_pruned_40.pt")
r["YOLO11 40% FP16"]    = benchmark("/home/jj7/Scalable/yolo11s_pruned_40.engine")
r["YOLO26 40% PyTorch"] = benchmark("/home/jj7/Scalable/YOLO26_pruned_40_fineTuned.pt")
r["YOLO26 40% FP16"]    = benchmark("/home/jj7/Scalable/YOLO26_pruned_40_fineTuned.engine")

print(f"\n{'Model':<25} {'Latency':>10} {'FPS':>8}")
print("="*45)
for name, ms in r.items():
    print(f"{name:<25} {ms:>8.2f}ms {1000/ms:>8.1f}")
```

```python
from ultralytics import YOLO

# Use full paths
model_original  = YOLO("/home/jj7/Scalable/yolo11s_pruned.pt")
model_quantized = YOLO("/home/jj7/Scalable/yolo11s_pruned.engine", task="detect")

# Validate on COCO128 (since you don't have your real dataset)
# This gives a rough idea of accuracy drop
results_original  = model_original.val(data="/tmp/calib_data.yaml",  imgsz=640, device=0, verbose=False)
results_quantized = model_quantized.val(data="/tmp/calib_data.yaml", imgsz=640, device=0, verbose=False)

print(f"Original  mAP50:    {results_original.box.map50:.4f}")
print(f"Quantized mAP50:    {results_quantized.box.map50:.4f}")
print(f"mAP50 drop:         {results_original.box.map50 - results_quantized.box.map50:.4f}")
print(f"Original  mAP50-95: {results_original.box.map:.4f}")
print(f"Quantized mAP50-95: {results_quantized.box.map:.4f}")
print(f"mAP50-95 drop:      {results_original.box.map - results_quantized.box.map:.4f}")
```
