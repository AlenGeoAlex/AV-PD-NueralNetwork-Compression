```python
from ultralytics import YOLO

# This will auto-download yolo11n.pt from Ultralytics
model = YOLO("yolo11n.pt")

# Train on YOUR dataset
results = model.train(
    data="/home/jj7/Scalable/Dataset/fixed_dataset.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,
    project="/home/jj7/Scalable/yolo11n_training",
    name="yolo11n_custom",
    verbose=True
)
print("Training done!")
```

```python
import zipfile, os

# Zip the YOLO11n trained model
zip_path = "/home/jj7/Scalable/yolo11n_custom.zip"

with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zf:
    zf.write(
        "/home/jj7/Scalable/yolo11n_training/yolo11n_custom/weights/best.pt",
        "yolo11n_custom_best.pt"
    )
    zf.write(
        "/home/jj7/Scalable/yolo11n_training/yolo11n_custom/weights/last.pt",
        "yolo11n_custom_last.pt"
    )

print(f"ZIP size: {os.path.getsize(zip_path)/1e6:.1f} MB")

from IPython.display import Javascript, display
display(Javascript("""
var a = document.createElement('a');
a.href = '/files/Scalable/yolo11n_custom.zip';
a.download = 'yolo11n_custom.zip';
document.body.appendChild(a);
a.click();
document.body.removeChild(a);
"""))
print("Download starting!")
```
