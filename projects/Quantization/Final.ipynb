```python
from ultralytics import YOLO

yaml = "/home/jj7/Scalable/Dataset/fixed_dataset.yaml"

models = {
    "YOLO26 20% PyTorch": "/home/jj7/Scalable/models/20pct/YOLO26_pruned_20_final.pt",
    "YOLO26 20% INT8":    "/home/jj7/Scalable/engines/20pct/YOLO26_pruned_20_final.engine",
    "YOLO26 40% PyTorch": "/home/jj7/Scalable/models/40pct/YOLO26_pruned_40_fineTuned.pt",
    "YOLO26 40% INT8":    "/home/jj7/Scalable/engines/40pct/YOLO26_pruned_40_fineTuned.engine",
}

results = {}
for name, path in models.items():
    print(f"Validating {name}...")
    m = YOLO(path, task="detect") if path.endswith(".engine") else YOLO(path)
    r = m.val(data=yaml, imgsz=640, device=0, verbose=False)
    results[name] = r.box.map50
    print(f"  mAP50: {r.box.map50:.4f}")

print("\n=== YOLO26 Accuracy Report ===")
print(f"{'Model':<25} {'mAP50':>8} {'Drop':>8} {'Status':>10}")
print("="*55)
for pct in ["20%", "40%"]:
    pt  = results[f"YOLO26 {pct} PyTorch"]
    eng = results[f"YOLO26 {pct} INT8"]
    drop = pt - eng
    status = "OK ✅" if drop < 0.01 else "WARN ⚠️" if drop < 0.03 else "FINE TUNE ❌"
    print(f"YOLO26 {pct} PyTorch    {pt:>8.4f} {'--':>8} {'baseline':>10}")
    print(f"YOLO26 {pct} INT8       {eng:>8.4f} {drop:>8.4f} {status:>10}")
```
